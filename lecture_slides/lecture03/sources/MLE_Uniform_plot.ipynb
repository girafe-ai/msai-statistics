```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
X = np.random.uniform(size=5)
```

```python
plt.axvline(X[0],ymax=0.2,color='black',linewidth=1.7,label='samples')

for x in X[1:]:
    plt.axvline(x,ymax=0.2,color='black',linewidth=1.7)

xs = np.linspace(max(X),1.5,50)
#plt.plot(np.linspace)
plt.plot(xs,(1/xs)**5,color='green',label='likelihood')
plt.axvline(x=max(X),ymin=0.2,ymax=0.95,color='green')
plt.plot(np.linspace(0,1,50),np.ones(50),color='red',linestyle='--',label='pdf')
plt.axvline(0,ymin=0,ymax=0.3,color='red',linestyle='--',alpha=0.9)
plt.axvline(1,ymin=0,ymax=0.3,color='red',linestyle='--',alpha=0.9)
#plt.plot(max(X),(1/max(X))**5)
plt.xlabel('x')
plt.legend()
#plt.plot()
plt.title('MLE for  $X\sim U(0,1)$')
plt.savefig('MLE_Uniform.png',dpi=1200)
```
