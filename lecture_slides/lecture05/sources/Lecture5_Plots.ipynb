```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
mean_vars = []
median_vars = []
midrange_vars = []

ns = [100,200,500,1000,2000,5000,10000,20000,50000,100000]

for n in ns:
    X = np.random.normal(loc=0,size=(n,1000))
    mean_vars.append(X.mean(axis=0).var())
    median_vars.append(np.median(X,axis=0).var())
    #midrange_vars.append(0.5*(X.min(axis=0)+X.max(axis=0)).var())
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)

ax1.plot(np.log10(ns),mean_vars,'--',color='red',label='mean')
ax1.plot(np.log10(ns),median_vars,'--',color='blue',label='median')

ax1.legend()

ax2.plot(np.log10(ns),np.log10(mean_vars),'.',color='red')
ax2.plot(np.log10(ns),np.log10(median_vars),'.',color='blue')

ax1.set_ylabel('Variance')

ax2.set_ylabel('log10(Variance)')
ax2.set_xlabel('log10(sample_size)')

fig.suptitle('$x\sim N(0,1)$, 1000 samples of size sample_size')

plt.savefig('Mean-vs-Median.png',dpi=1200)
```

```python
from scipy.stats import norm
```

```python
xs = np.linspace(-5,5,100)

plt.plot(xs, 0.5*(norm.pdf(xs,loc=-2,scale=0.4)+norm.pdf(xs,loc=2,scale=0.9)),color='black',
        label='$0.5(p_1(x)+p_2(x))$')

plt.plot(xs, norm.pdf(xs,loc=-2,scale=0.4),alpha=0.5,color='red',linestyle='--',label='$\mu=-2,\sigma=0.6$')
plt.plot(xs, norm.pdf(xs,loc=2,scale=0.9),alpha=0.5,color='green',linestyle='--',label='$\mu=2,\sigma=0.9$')

plt.title('Gaussian mixture model')
plt.xlabel('$x$')

plt.legend()

plt.savefig('GaussianMixture.png',dpi=1200)
```
